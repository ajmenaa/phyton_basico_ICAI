Calcular Datos

```python
tipo_cambio_maximo = tipo_cambio['Compra'].max()
tipo_cambio_maximo
```

```python
entidad_con_maximo = tipo_cambio[tipo_cambio['Compra'] == tipo_cambio_maximo]['Entidad']
entidad_con_maximo.to_list()
```

```python
tipo_cambio_minimo = tipo_cambio['Venta'].min()
tipo_cambio_minimo
```

```python
entidad_con_minimo = tipo_cambio[tipo_cambio['Venta'] == tipo_cambio_minimo]['Entidad']
entidad_con_minimo.to_list()
```

```python
#Cambiar tipo para cálculo
promedio_venta = tipo_cambio['Venta'].astype(float).mean()
print(promedio_venta)
```

```python
#Seleccionar columnas
columnas_numericas = ['Compra', 'Venta', 'Diferencial']
tipo_cambio[columnas_numericas] = tipo_cambio[columnas_numericas].apply(pd.to_numeric, errors='coerce')
```

```python
promedio_venta = tipo_cambio['Venta'].mean()
promedio_venta
```

```python
promedio_venta_tipo = tipo_cambio.groupby('Tipo')['Venta'].mean()
promedio_venta_tipo
```

```python
promedio_venta_tipo.sort_values()
```

```python
promedio_venta_tipo.sort_values(ascending = False)
```
