```python
import pandas as pd

# Realizar el web scrapping

df  = pd.read_html('https://gee.bccr.fi.cr/IndicadoresEconomicos/Cuadros/frmConsultaTCVentanilla.aspx', encoding='utf-8', decimal=',' , thousands='.')
```

```python
df[2]
```

```python
# Dimensiones DataFrame
df[2].shape
```

```python
# Guardamos el DataFrame
tipo_cambio = df[2].copy()
del(df)

tipo_cambio.head()
```

```python
tipo_cambio.columns
```

Extraer la primera fila del DataFrame, y asignarle a la columna la primera fila

```python
primera_fila = tipo_cambio.iloc[0] #Extraer fila 0
```

```python
# prompt: Renombrar las columnas con la primera_fila
# tipo_cambio.rename(primera_fila,axis=1, inplace=1) # Opcion 2
tipo_cambio.columns = primera_fila
tipo_cambio.head()
```

```python
tipo_cambio.drop(0, inplace=True) # Eliminar la primera fila ---- OJO con el inplace
```

```python
# Renombramos la primera columna para que sea más fácil consultarla
tipo_cambio.rename(columns={'Tipo de Entidad':'Tipo','Entidad Autorizada':'Entidad','Diferencial Cambiario':'Diferencial'}, inplace=True)

tipo_cambio.columns
```

```python
#Eliminar Nulos
tipo_cambio.dropna()
```

```python
#Rellenar nulos
tipo_cambio['Tipo'].fillna('No VALOR').tail()
```

```python
# Remplazar valores NaN con el último valor no nulo en la misma columna
tipo_cambio['Tipo'].ffill(inplace=True)

tipo_cambio
```

```python
len(tipo_cambio)
```

```python
#Eliminar la ultima fila del DataFrmae tipo_cambio
#tipo_cambio.drop(tipo_cambio.tail(1).index)
#tipo_cambio.drop(len(tipo_cambio))
#tipo_cambio.dropna(subset=['Diferencial'])
#tipo_cambio = tipo_cambio[:-1].copy()


tipo_cambio.drop(tipo_cambio.tail(1).index , inplace = True)
```

Calcular Datos

```python
# Consultar la entidad con la compra mas alta
tipo_cambio_maximo = tipo_cambio['Compra'].max()
entidades_con_maximo = tipo_cambio[tipo_cambio['Compra'] == tipo_cambio_maximo]['Entidad']


print('Tipo de cambio máximo: ',tipo_cambio_maximo)
print('Entidades con tipo de cambio máximo: \n',entidades_con_maximo)
```

```python
# Consultar la entidad con la venta mas baja
tipo_cambio_minimo = tipo_cambio['Venta'].min()
entidades_con_maximo = tipo_cambio[tipo_cambio['Venta'] == tipo_cambio_minimo]['Entidad']


print('Tipo de cambio máximo: ',tipo_cambio_minimo)
print('Entidades con tipo de cambio máximo: \n',entidades_con_maximo)
```

```python
# Selecionar columnas
columnas_numericas = ['Compra','Venta','Diferencial']
tipo_cambio[columnas_numericas] = tipo_cambio[columnas_numericas].apply(pd.to_numeric,errors='coerce')
```
